## Step 2: Load a FREE model

````python
from transformers import pipeline
generator = pipeline("text2text-generation", model="google/flan-t5-base")
````

## Step 3: Create a Search Tool (FREE)

```
1. You are importing the DDGS class from the duckduckgo_search library
This library lets you perform search queries (like Google) using DuckDuckGo, but programmatically (via Python)
No API key needed → good for free agents

```


````python
from ddgs import DDGS
````



```
2. You are defining a function (tool) named search_tool
It takes one input:
query → the search text (e.g., "latest AI news")

👉 In an AI agent, this becomes a tool the agent can call

3. You initialize an empty string
This will store all search results combined

4. You create a DuckDuckGo search session using DDGS()
with ensures:
automatic cleanup
safe resource handling

👉 Think of it like opening & closing a connection

5. You perform a text search
Parameters:
query → what user asked
max_results=3 → only fetch top 3 results

👉 Output:
A list of dictionaries like:

```



````python
def search_tool(query):
    results_text = ""
    with DDGS() as ddgs:
        results = ddgs.text(query, max_results=3)
        for r in results:
            print("href",r["href"])
            results_text +=r["body"] + "\n"
    return results_text if results_text else "No results found"
````

## Step 4: Build a CLEAR Agent (important part)

````python
def agent(query):
    print("User:", query)

    # Step 1: THINK
    if "latest" in query or "news" in query or "today" in query:
        print("🧠 Thinking: I need to search the internet")

        # Step 2: ACT
        search_results = search_tool(query)
        print("🔎 Results:\n", search_results[:200], "\n")

        prompt = f"""
        Based on the information below, give a clear summary.

        Info:
        {search_results}

        Question: {query}
        Answer in simple words:
        """
    else:
        print("🧠 Thinking: I can answer directly")
        prompt = f"Answer this clearly: {query}"

    # Step 3: RESPOND
    output = generator(prompt, max_length=300, num_return_sequences=1)
    answer = output[0]["generated_text"]
    print("💬 Answer:\n", answer)

    return answer
````
